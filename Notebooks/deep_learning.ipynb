## 1. Classification with a Fully Convolutional Network (FCN)

For classification, the objective is to assign each input time series to one of a finite number of classes. The example below loads a benchmark classification dataset, trains a Fully Convolutional Network, and produces class probability estimates for the test set.

We use the `ArrowHead` dataset from the [UCR archive](https://www.timeseriesclassification.com), which is downloaded automatically by aeon.

```python
from aeon.datasets import load_classification
from aeon.classification.deep_learning import FCNClassifier

X_train, y_train = load_classification("ArrowHead", split="TRAIN")
X_test, y_test = load_classification("ArrowHead", split="TEST")

print("Train shape:", X_train.shape, "| Test shape:", X_test.shape)

clf = FCNClassifier(n_epochs=50, batch_size=16)
clf.fit(X_train, y_train)

probs = clf.predict_proba(X_test)
probs[:5]
```

```python
from sklearn.metrics import accuracy_score

y_pred = clf.predict(X_test)
print("Test accuracy:", accuracy_score(y_test, y_pred))
```

## 2. Regression with a Residual Network (ResNet)

For regression, the target variable is continuous rather than categorical. The workflow remains the same, but the model outputs numerical predictions rather than class labels or probabilities.

Here we use the `BarCrawl6min` dataset from the time series extrinsic regression archive, training a Residual Network.

```python
from aeon.datasets import load_regression
from aeon.regression.deep_learning import ResNetRegressor

X_train, y_train = load_regression("BarCrawl6min", split="TRAIN")
X_test, y_test = load_regression("BarCrawl6min", split="TEST")

print("Train shape:", X_train.shape, "| Test shape:", X_test.shape)

reg = ResNetRegressor(n_epochs=50, batch_size=16)
reg.fit(X_train, y_train)

preds = reg.predict(X_test)
preds[:5]
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error

rmse = np.sqrt(mean_squared_error(y_test, preds))
print("Test RMSE:", rmse)
```

## 3. Going further: multivariate series and rehabilitation

The deep learning estimators in aeon natively support **multivariate** time series, where each series has several channels (e.g. joint coordinates over time). The same `fit`/`predict` interface applies.

In the final part of the tutorial we consider **skeleton-based human motion rehabilitation** using the [Rehab-Pile](https://arxiv.org/abs/2507.21018) archive, a collection of rehabilitation exercise recordings designed for movement assessment. Skeleton sequences are represented as multivariate time series — with each dimension corresponding to joint positions or angles — and used with deep learning models in aeon to recognise exercise types, assess execution quality, or predict continuous rehabilitation scores.

Other architectures available in aeon worth experimenting with include `InceptionTimeClassifier`, `LITETimeClassifier`, and their regression counterparts. See the [aeon API reference](https://www.aeon-toolkit.org/en/stable/api_reference.html) for the full list.
